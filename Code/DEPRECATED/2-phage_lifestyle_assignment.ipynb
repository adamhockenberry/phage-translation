# Compile data into a dataframe
I really feel like I have another dataframe laying around that this data should be added to

```python
full_listy = []
for host_dir in glob.glob('../Data/MVP_data/host_linked_genomes/*_rep_viruses/'):
    print('#####', host_dir)
    host_id = host_dir.split('/')[-2].split('_')[0]
    for bacphlip_file in glob.glob(host_dir + '*.fasta.bacphlip'):
        phage_id = bacphlip_file.split('/')[-1].split('.')[0]
        temp_df = pd.read_csv(bacphlip_file, '\t')
        prob_diff = abs(temp_df.loc[0]['Virulent']-temp_df.loc[0]['Temperate'])
        lifestyle = ''
        if temp_df.loc[0]['Virulent']>temp_df.loc[0]['Temperate']:
            lifestyle = 'Virulent'
        else:
            lifestyle = 'Temperate'
        full_listy.append((host_id, phage_id, lifestyle, prob_diff))
```

```python
df = pd.DataFrame(full_listy, columns=['host_id', 'phage_id', 'phage_lifestyle', 'lifestyle_probability_difference'])
df.head()
```

```python
df.to_csv('../Data/MVP_data/lifestyle_results_BACPHLIP.tsv', sep='\t')
```
